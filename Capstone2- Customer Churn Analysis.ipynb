# Data Preprocessing for Model Fitting

```python
churn = pd.read_csv("C:/Users/umalviya/Desktop/Churn Modeling.csv")
churn = churn.drop(["Surname","CustomerId","RowNumber"], axis=1)
print(churn)
f,ax = plt.subplots(figsize=(10, 8))
sns.heatmap(churn.iloc[:,:10].corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

### Converting categorical data into numerical for model fitting

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
churn['gender'] = le.fit_transform(churn['Gender'])
churn['geography']=le.fit_transform(churn['Geography'])
churn
```

```python
churn['Gender']=churn['Gender'].astype('category')
churn['Geography']=churn['Geography'].astype('category')
churn.dtypes
```

```python
churn['Gender']=churn['Gender'].cat.codes
churn['Geography']=churn['Geography'].cat.codes
churn.drop(['geography','gender'],axis=1,inplace=True)
churn
```

### Splitting the dataset for modelling

```python
Y = churn.Exited
X = churn.drop(['Exited'], axis=1)
print(y[0:5])
print(X[0:5])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=42)

print('Training Set Shape:   ', X_train.shape, Y_train.shape)
print('Test Set Shape:       ', X_test.shape, Y_test.shape)
```

# Prediction using Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import accuracy_score 
logreg=LogisticRegression()
logreg.fit(X_train,Y_train)
```

```python
# Train the logistic regression model
logreg=LogisticRegression()
logreg.fit(X_train,Y_train)

# Predict on training dataset
lgpred_train = logreg.predict(X_train)
print("Model accuracy on training dataset: ",metrics.accuracy_score(Y_train, lgpred_train))
print('================================================================')

#Predict on test dataset
lgpred_test = logreg.predict(X_test)
print("Model accuracy on test dataset: ",metrics.accuracy_score(Y_test, lgpred_test))
print('================================================================')

conf_matlg=confusion_matrix(Y_test, lgpred_test, labels=None, sample_weight=None)
print("Confustion Matrix: ")
print(conf_matlg)
print('================================================================')

print(metrics.classification_report(Y_test, lgpred_test))
```

```python
# Plotting Confusion matrix
names = np.unique(lgpred_test)
sns.heatmap(conf_matlg, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=names, yticklabels=names)
plt.xlabel('Truth')
plt.ylabel('Predicted')
conf_matlg
```

```python
# Coefficients :
weights = pd.Series(logreg.coef_[0],
                 index=X.columns.values)
weights.sort_values()
```

#### With logistic regression, we predicted the customers who are likely to churn with 80% accuracy.According to logistic regression age,geography and balance are more important variable
